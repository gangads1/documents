```python
from pymongo import MongoClient
import pandas as pd
```

```python
df1 = pd.read_csv("..\data\data_1.csv", index_col=False, delimiter=",")
df1.head()
```

```python
df2 = pd.read_csv('../data/data_2_2.csv')
df3 = pd.read_csv('../data/data_3.csv')
df4 = pd.read_csv('../data/data_4_2.csv')
```

### Connecting to MongoDB, creating iris_project database and push dataframe into MongoDB collection

#### Data_1

```python
client = MongoClient("mongodb://localhost:27017")

db = client["Iris_project"]

data_1 = df1.to_dict("records")

collection = db["data_1"]
collection.insert_many(data_1)
cursor = collection.find({})
for document in cursor:
    print(document)
    
client.close()
```

### Retrieving data from MongoDB to Dataframe

```python
client = MongoClient("mongodb://localhost:27017")

db = client["Iris_project"]


collection = db["data_1"]
cursor = collection.find({})
data_set_1 = pd.DataFrame(list(cursor))

data_set_1
```

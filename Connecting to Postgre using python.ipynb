```python
import psycopg2
```

### Python to Postgre SQL connection

```python
# Establish a connection to the PostgreSQL database
conn = psycopg2.connect(
    host="localhost",
    database="iris",
    user="postgres",
    password="root"
)

# Create a cursor object to execute SQL statements
cursor = conn.cursor()
```

### table creation

```python
# Define the table name and column names with their respective data types
table_name = 'data_1'
column1_name = 'Id'
column1_type = 'INTEGER'
column2_name = 'PetalLengthCm'
column2_type = 'FLOAT'
column3_name = 'PetalWidthCm'
column3_type = 'FLOAT'

# Create the table
create_table_query = f'''
    CREATE TABLE {table_name} (
        {column1_name} {column1_type},
        {column2_name} {column2_type},
        {column3_name} {column3_type}
    )
'''
cursor.execute(create_table_query)

# Commit the changes
conn.commit()
```

### load the dataset and create dataframe

```python
import pandas as pd

df1 = pd.read_csv('../data/data_1.csv')
```

```python
df1.head()
```

### insert the dataframe to created table

```python
for row in df1.itertuples(index=False):
    insert_query = f'''
        INSERT INTO {table_name} (Id, PetalLengthCm, PetalWidthCm)
        VALUES (%s, %s, %s)
    '''
    cursor.execute(insert_query, row)
conn.commit()
```

### check the rows of the data table

```python
cursor = conn.cursor()
# Define the table name
table_name = 'data_1'

# Select all data from the table
select_query = f"SELECT * FROM {table_name}"
cursor.execute(select_query)

# Fetch all the rows
rows = cursor.fetchall()

# Print the retrieved data
for row in rows:
    print(row)
```

### again the load the dataset fom postgreSQL

```python
import warnings
warnings.filterwarnings(action='ignore')


load_df = pd.read_sql(f"SELECT * FROM {table_name}", conn)
```
